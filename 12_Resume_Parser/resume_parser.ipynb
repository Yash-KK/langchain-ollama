## Load PDF Resume Data

```python
from pathlib import Path

from langchain_pymupdf4llm import PyMuPDF4LLMLoader

path = Path("/home/yash/Desktop/Code/1. Langchain and Ollama/12_Resume_Parser/resumes/Yash_Kharche_Resume.pdf")
loader = PyMuPDF4LLMLoader(path, mode="page", table_strategy = "lines")
docs = loader.load()
```

```python
docs[0].metadata
resume_content = docs[0].page_content
```

```python
parsed_resume = ask_llm(resume_content)
```

```python
parsed_resume
```

```python
parsed_validated_json = validate_json(parsed_resume)
```

```python
parsed_validated_json
```

```python
import json

filename = path.stem
output_file = filename.replace('.pdf', '.json')
output_file = 'parsed_resumes/{}'.format(output_file)

json.dump(parsed_validated_json, open(output_file, 'w'), indent=4)
```
