```python
from dotenv import load_dotenv

from langfuse import get_client
from langfuse.langchain import CallbackHandler

load_dotenv()
langfuse = get_client()
langfuse_handler = CallbackHandler()
```

```python
from langchain_ollama import ChatOllama
from helpers.common import (
  BASE_URL,
  # DEEPSEEK_MODEL,
  QWEN_MODEL
)
```

```python
llm = ChatOllama(
  model=QWEN_MODEL,
  base_url=BASE_URL,
  validate_model_on_init=True
  
)
```

```python
input_text = "Hello, How are you doing?"
async for chunk in llm.astream(input_text, config={"callbacks": [langfuse_handler], "run_name": "ollama-simple-chat"}):
  print(chunk.content, end="")
```

## Langchain Messages

```python
from langchain.messages import SystemMessage, HumanMessage

system_message = SystemMessage("You are an elementary Teacher. Answer in short points.")
human_input = HumanMessage("Tell me about earth in 3 points,")

messages = [system_message, human_input]
async for chunk in llm.astream(messages, config={"callbacks": [langfuse_handler], "run_name": "ollama-messages-elementary"}):
  print(chunk.content, end="")
```

## Langchain Prompt Templates

```python
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate

system_prompt = SystemMessagePromptTemplate.from_template("You are a {level} teacher. Answer in short points.")
human_prompt = HumanMessagePromptTemplate.from_template("Tell me about {topic} in {number} points.")
messages = [system_prompt, human_prompt]
chat_prompt = ChatPromptTemplate(messages)

```

```python
final_prompt = chat_prompt.invoke({
  "level": "science",
  "topic": "soil",
  "number": 5
})
```

```python
async for message in llm.astream(final_prompt, config={"callbacks": [langfuse_handler], "run_name": "ollama-prompt-template"}):
  print(message.content, end="")
```
